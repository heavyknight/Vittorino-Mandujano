## 2. RegresiÃ³n Lineal

```python
variables=['LotFrontage', 'LotArea', 'OverallQual', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'TotalBsmtSF', 'BsmtUnfSF', '1stFlrSF', '2ndFlrSF', 'GrLivArea', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF', 'MoSold', 'YrSold']
X=data[variables]

from sklearn.preprocessing import Imputer

imputer = Imputer(strategy="median")
imputer.fit(X)
Xx = imputer.transform(X)
Xtraining = pd.DataFrame(Xx, columns=X.columns,index = list(X.index.values))

Xtraining.info()
```

### 2.1 Entrenamiento
Se hallarÃ¡n los valores de $\theta$ del ajuste de curva

```python
y=data['SalePrice']
y=y.as_matrix(columns=None)
X=Xtraining.as_matrix(columns=None)
Q=[[1]]
for i in range(1459):
    Q=Q+[[1]]
Q=np.asarray(Q)
X=np.hstack((Q,X))
Xx=np.matmul(X.transpose(),X)
Xx=np.linalg.inv(Xx)
B=np.matmul(Xx,np.matmul(X.transpose(),y)) #Se usÃ³ el mÃ©todo matricial
B
```

### 2.2 Testeo

```python
y_test=np.matmul(X,B)
```

CÃ¡lculo del RMSE

```python
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y, y_test)))
```

```python
testdata=pd.read_csv('test.csv',index_col=0)
```

Calculo de predicciones en base al archivo "test" de Kaggle

```python
Q=[[1]]
for i in range(1458):
    Q=Q+[[1]]
Q=np.asarray(Q)
Xprueba=np.hstack((Q,Xprueba))
y_pred =np.matmul(Xprueba,B)
```
